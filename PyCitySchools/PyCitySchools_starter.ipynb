```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python

school_data_complete['passing reading'] = school_data_complete['reading_score'] >= 70
school_data_complete['passing math'] = school_data_complete['math_score'] >= 70
school_data_complete['passing overall'] = (school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)

school_data_complete.head()
```

```python
summary = pd.DataFrame(
    {'Total Schools' : [school_data_complete['school_name'].nunique()],
     'Total Students' : [school_data_complete['student_name'].count()],
     'Total Budget' : [school_data['budget'].sum()],
     'Average Math Score' : [school_data_complete['math_score'].mean()],
     'Average Reading Score' : [school_data_complete['reading_score'].mean()],
     '% Passing Math' : [school_data_complete.loc[school_data_complete['math_score'] >= 70, :].count()['student_name']/school_data_complete['student_name'].count()],
     '% Passing Reading' : [school_data_complete.loc[school_data_complete['reading_score'] >= 70, :].count()['student_name']/school_data_complete['student_name'].count()],
     '% Overall Passing' : [school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70), :].count()['student_name']/school_data_complete['student_name'].count()]
    }
)
summary.style.format({'Total Students' : '{:,}',
                    'Total Budget' : '${:20,.0f}',
                     '% Passing Math' : '{:.2%}',
                     '% Passing Reading' : '{:.2%}',
                     '% Overall Passing' : '{:.2%}'
                     })
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# define object as df.groupby schools then use print method ***see groupby demo use mapping to format whole df after calcs
gbs = school_data_complete.groupby('school_name')
school_summary = pd.DataFrame({'School Type' : gbs['type'].first(),
                               'Total Students' : gbs['student_name'].count(),
                               'Total School Budget' : gbs['budget'].first(),
                               'Per Student Budget' : gbs['budget'].first()/gbs['student_name'].count(),
                               'Average Math Score' : gbs['math_score'].mean(),
                               'Average Reading Score' : gbs['reading_score'].mean(),
                               '% Passing Math' : gbs['passing math'].sum()/gbs['passing math'].count(),
                               '% Passing Reading' : gbs['passing reading'].sum()/gbs['passing reading'].count(),
                               '% Passing Overall' : gbs['passing overall'].sum()/gbs['passing overall'].count()
                              })

school_summary.style.format({'Total Students' : '{:,}',
                             'Per Student Budget' : '${:20,.0f}',
                             'Total School Budget' : '${:20,.0f}',
                             '% Passing Overall' : '{:.2%}',
                             '% Passing Reading' : '{:.2%}',
                             '% Passing Math' : '{:.2%}'})
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top5 = school_summary.nlargest(5,'% Passing Overall')
top5.style.format({'Total Students' : '{:,}',
                   'Per Student Budget' : '${:20,.0f}',
                   'Total School Budget' : '${:20,.0f}',
                   '% Passing Overall' : '{:.2%}',
                   '% Passing Reading' : '{:.2%}',
                   '% Passing Math' : '{:.2%}'})
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
nineth = school_data_complete.loc[school_data_complete['grade'] == '9th', :]
tenth = school_data_complete.loc[school_data_complete['grade'] == '10th', :]
eleventh = school_data_complete.loc[school_data_complete['grade'] == '11th', :]
twelfth = school_data_complete.loc[school_data_complete['grade'] == '12th', :]

avg9 = nineth.groupby('school_name')
avg10 = tenth.groupby('school_name')
avg11 = eleventh.groupby('school_name')
avg12 = twelfth.groupby('school_name')

math_by_grade = pd.DataFrame({'9th' : avg9['math_score'].mean(),
                              '10th' : avg10['math_score'].mean(),
                              '11th' : avg11['math_score'].mean(),
                              '12th' : avg12['math_score'].mean()
                             })

math_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
reading_by_grade = pd.DataFrame({'9th' : avg9['reading_score'].mean(),
                                 '10th' : avg10['reading_score'].mean(),
                                 '11th' : avg11['reading_score'].mean(),
                                 '12th' : avg12['reading_score'].mean()
                               })

reading_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#binning by spending
spending_bins = [0, 584, 629, 644, 675]
spending_bin_names = ['<$584', '$585-629', '$630-644', '$645-675']
school_summary['spending bin'] = pd.cut(school_summary['Per Student Budget'], spending_bins, labels=spending_bin_names)
gbsb = school_summary.groupby('spending bin')

outcomes_by_spending = pd.DataFrame({'Average Math Score' : gbsb['Average Math Score'].mean()/100,
                                     'Average Reading Score' : gbsb['Average Reading Score'].mean()/100,
                                     '% Passing Math' : gbsb['% Passing Math'].mean(),
                                     '% Passing Reading' : gbsb['% Passing Reading'].mean(),
                                     '% Passing Overall' : gbsb['% Passing Overall'].mean()
                                    })

outcomes_by_spending.style.format({'Average Math Score' : '{:.2%}',
                                   'Average Reading Score' : '{:.2%}',
                                   '% Passing Overall' : '{:.2%}',
                                   '% Passing Reading' : '{:.2%}',
                                   '% Passing Math' : '{:.2%}'})
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
size_bin_names = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']
school_summary['size bin'] = pd.cut(school_summary['Total Students'], size_bins, labels=size_bin_names)
gbzb = school_summary.groupby('size bin')

outcomes_by_size = pd.DataFrame({'Average Math Score' : gbzb['Average Math Score'].mean()/100,
                                     'Average Reading Score' : gbzb['Average Reading Score'].mean()/100,
                                     '% Passing Math' : gbzb['% Passing Math'].mean(),
                                     '% Passing Reading' : gbzb['% Passing Reading'].mean(),
                                     '% Passing Overall' : gbzb['% Passing Overall'].mean()
                                    })

outcomes_by_size.style.format({'Average Math Score' : '{:.2%}',
                                   'Average Reading Score' : '{:.2%}',
                                   '% Passing Overall' : '{:.2%}',
                                   '% Passing Reading' : '{:.2%}',
                                   '% Passing Math' : '{:.2%}'})
```
